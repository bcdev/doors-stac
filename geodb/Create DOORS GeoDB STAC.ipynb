```python
def create_stac_collection(collection_id, collection_title, provider_name, extent, summaries, file_name):
    providers = [
        {"name": provider_name, "roles": ["producer"]}
        # {"name": provider_name}
    ]        
    links = [
        {"rel": "self", "href": file_name, "type": "application/json"},
        {"rel": "parent", "href": "catalog.json", "type": "application/json"},
    ]
    collection = {
        "stac_version": "1.0.0",
        "id": collection_id,
        "title": collection_title,
        "description": collection_title,
        "providers": providers,
        "extent": extent,
        "links": links,
        "summaries": summaries
    }
    with open(file_name, "w") as f:
        json.dump(collection, f, indent=2)
```

```python
def _create_id(title, ids):
    identity = title.lower().replace(" ", "_")
    i = 1
    first = True
    while(identity in ids):
        if first:
            identity = identity + "_"
        identity = identity + f"{i}"
        i += 1
    return identity
```

```python
def _create_extent(gdf):
    extent = {}
    # geometry = gdf.geometry.unary_union.envelope.wkt    
    geometry_bounds = gdf.geometry.unary_union.envelope.bounds
    extent["spatial"] = {
        "bbox": [[geometry_bounds[0], geometry_bounds[1], geometry_bounds[2], geometry_bounds[3]]]
    }
    if "timestamp" in gdf.columns:
        min_time = pd.to_datetime(gdf.timestamp.min()).strftime('%Y-%m-%dT%H:%M:%S')
        max_time = pd.to_datetime(gdf.timestamp.max()).strftime('%Y-%m-%dT%H:%M:%S')
        extent["temporal"] = {
            "interval": [[min_time, max_time]]
        }
    return extent
```

```python
def _create_summaries(gdf):
    summaries = {}
    for column_name in gdf.columns:
        if column_name == "id":
            continue
        if gdf[column_name].dtype == "float" or gdf[column_name].dtype == "int":        
            summaries[column_name] = {
                "minimum": str(gdf[column_name].min()),
                "maximum": str(gdf[column_name].max()),
            }
    return summaries
```

```python
catalog = {
    "stac_version": "1.0.0",
    "id": "doors-geodb-catalog",
    "title": "DOORS GeoDB Catalog",
    "description": "A catalog providing an overview about the DOORS GeoDB Data"
}

links = [
    {"rel": "self", "href": "catalog.json", "type": "application/json"},
    {"rel": "root", "href": "catalog.json", "type": "application/json"}
]

ids = []
for dfm in df_metadata:
    collections = df[df["owner"].str.lower() == dfm.get("owner")]
    # print(collections.collection)
    for i, collection in collections.iterrows():
        print(f"{i} of {len(collections)}")
        collection_id = _create_id(collection.collection, ids)
        file_name = f"{collection_id}.json"

        if not os.path.exists(file_name):
            # print(f"Create {file_name}")
            try:
                num_rows = geodb.count_collection_rows(collection.collection, database=collection.database)
                if num_rows > 1000000:
                    gdf = geodb.get_collection_pg(
                        collection.collection, where='id % 100 = 0', database=collection.database
                    )
                else:
                    gdf = geodb.get_collection(collection.collection, database=collection.database)
            except:
                print(f"Could not access collection {collection_id}")
                continue
            
            extent = _create_extent(gdf)
            summaries = _create_summaries(gdf)

            create_stac_collection(collection_id, collection.collection, dfm.get("organisation"), extent, summaries, file_name)

        links.append(
            {"rel": "child", "href": file_name, "type": "application/json", "title": collection.collection}
        )
        # if i > 0:
            # break
    # break
catalog["links"] = links

with open("catalog.json", "w") as f:
    json.dump(catalog, f, indent=2)
```
